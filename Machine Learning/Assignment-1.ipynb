```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
data = {
    'City': ['Pune', 'Mumbai', 'Delhi', 'Chennai'],
    'Temperature': [28, 32, 35, 34],
    'Humidity': [60, 70, 55, 65]
}

df_created = pd.DataFrame(data)
print("Created Dataset:")
print(df_created)
```

```python
df = pd.read_csv("temperature.csv")

print("\nLoaded Dataset:")
print(df)
```

```python
iris = datasets.load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

print("\nSklearn Dataset (Iris):")
print(df_iris.head())
```

```python
df.replace("NA", np.nan, inplace=True)
```

```python
print("\nStatistical Measures:")

print("\nMean:")
print(df.mean(numeric_only=True))

print("\nMedian:")
print(df.median(numeric_only=True))

print("\nMode:")
print(df.mode())

print("\nVariance:")
print(df.var(numeric_only=True))

print("\nStandard Deviation:")
print(df.std(numeric_only=True))
```

```python
reshaped_data = df.values.reshape(-1, len(df.columns))
print("\nReshaped Data:")
print(reshaped_data)
```

```python
extra_data = pd.DataFrame({
    'City': ['Pune', 'Mumbai', 'Delhi', 'Chennai'],
    'Rainfall': [120, 200, 80, 150]
})

merged_data = pd.merge(df, extra_data, on='City')
print("\nMerged Dataset:")
print(merged_data)
```

```python
print("\nMissing Values Before Handling:")
print(df.isnull().sum())

# Fill missing values with column mean
df_filled = df.fillna(df.mean(numeric_only=True))

print("\nDataset After Handling Missing Values:")
print(df_filled)

```

```python
normalized_df = df.copy()

for column in normalized_df.select_dtypes(include=np.number):
    normalized_df[column] = (
        normalized_df[column] - normalized_df[column].min()
    ) / (
        normalized_df[column].max() - normalized_df[column].min()
    )

print("\nMin-Max Normalized Dataset:")
print(normalized_df)
```
