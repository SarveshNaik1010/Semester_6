```python
import pandas as pd;
import numpy as np;
```

```python
df = pd.read_csv("temperature.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.replace("NA", np.nan, inplace=True)
```

```python
df["AverageTemperatureFahr"] = pd.to_numeric(df["AverageTemperatureFahr"])
df["AverageTemperatureUncertaintyFahr"] = pd.to_numeric(
    df["AverageTemperatureUncertaintyFahr"]
)

df.head(20)
```

```python
mean_temp = df["AverageTemperatureFahr"].mean()
mean_temp
```

```python
median_temp = df["AverageTemperatureFahr"].median()
median_temp
```

```python
min_temp = df["AverageTemperatureFahr"].min()
max_temp = df["AverageTemperatureFahr"].max()

min_temp, max_temp
```
